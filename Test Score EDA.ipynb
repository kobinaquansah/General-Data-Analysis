```python
def filtercondition(f):
    padf=df[["pa","ma","re","wr"]][df["ge"]==gedict[f]]
    educond=[padf["pa"]==padict[i] for i in range (6)]
    eduscores=[padf[educond[i]].iloc[:,[1,2,3]].apply(np.mean) for i in range (6)]
    mean_scores=list(map(lambda x: x.mean(), eduscores))
    ma_score=list(map(lambda x: x[0], eduscores))
    re_score=list(map(lambda x: x[1], eduscores))
    wr_score=list(map(lambda x: x[2], eduscores))

    global edumeans,eduma,edure,eduwr
    global tot_mean,math_mean,read_mean,write_mean
    edumeans=[[diction[2][i],mean_scores[i]] for i in range (6)]
    eduma=[[diction[2][i],ma_score[i]] for i in range (6)]
    edure=[[diction[2][i],re_score[i]] for i in range (6)]
    eduwr=[[diction[2][i],wr_score[i]] for i in range (6)]
    

    tot_mean=pd.DataFrame(edumeans).replace({"bachelor's degree":5,
                                             "some college":4,
                                             "master's degree":6,
                                             "associate's degree":3,
                                             "high school":2,
                                             "some high school":1})

    math_mean=pd.DataFrame(eduma).replace({"bachelor's degree":5,
                                        "some college":4,
                                        "master's degree":6,
                                        "associate's degree":3,
                                        "high school":2,
                                        "some high school":1})

    read_mean=pd.DataFrame(edure).replace({"bachelor's degree":5,
                                        "some college":4,
                                        "master's degree":6,
                                        "associate's degree":3,
                                        "high school":2,
                                        "some high school":1})

    write_mean=pd.DataFrame(eduwr).replace({"bachelor's degree":5,
                                        "some college":4,
                                        "master's degree":6,
                                        "associate's degree":3,
                                        "high school":2,
                                        "some high school":1})
    
    global a
    a=['tot_mean','math_mean','read_mean','write_mean']
    def columnmaker(x):
        eval(x).columns=([eval('x'),1])
        eval(x).sort_values(eval(x).columns[0], inplace=True)


     
    list(map(lambda u: columnmaker(u),a))

    
    a=[tot_mean,math_mean,read_mean,write_mean]

    plt.figure(figsize=(15,10))
    list(map(lambda u: plt.scatter(x=u.iloc[:,0].replace({
                                        1:"some high school",
                                        2:"high school",
                                        3:"associate's degree",
                                        4:"some college",
                                        5:"bachelor's degree",
                                        6:"master's degree"
                                        }), y=u[1], label=u.columns[0]),a))
    plt.legend()
    
    plt.show()

    def findcorrcoef(x):
        return np.corrcoef(np.transpose(x))

    global educorrcoef

    educorrcoef=list(map(lambda x:findcorrcoef(x), a))
    for i in range (len(educorrcoef)):
        educorrcoef[i]=educorrcoef[i][0,1]
    global subjectlist
    subjectlist=["mean","math","reading","writing"]
    educorrcoef
    finaldf=[subjectlist,educorrcoef]
    finaldf=pd.DataFrame(finaldf)
    print(finaldf)
    return(educorrcoef)
```

```python
girls=filtercondition(0)
```

IN THE SAMPLE OF FEMALES IN THIS STUDY: 
1. Overall:Parental educational achievement accounts for 91 percent of mean score varation.
2. Math skills are less correlated with parental educational background than reading and writing skills.
i.e. Parental educational level is a better predictor of language skills than mathematical ability in the females IN THIS STUDY.


```python
boys=filtercondition(1)
```

IN THE SAMPLE OF BOYS IN THIS STUDY, 
1. Overall: Parental educational achievement accounts for 84% of the mean score varation
2. Writing skills are more correlated with parental educational achievement than math and reading skills.
i.e. Parental educational level is a better predictor of writing skills than math and reading skills.
